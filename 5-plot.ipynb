```python
import pickle
import matplotlib.pyplot as plt
files_acc = ['alex_sgd_0.01_train_acc.txt',
        'alex_sgd_0.01_val_acc.txt']
files_loss = ['alex_sgd_0.01_train_loss.txt',
        'alex_sgd_0.01_val_loss.txt']
def load(path):
    with open(path, 'rb') as handle:
        b = pickle.load(handle)
    return b
```

```python
plt.style.use('seaborn-whitegrid')
fig,ax = plt.subplots(figsize=(16,9))
data = load(f'current_model/alex_sgd_0.01_train_loss.txt')
ax.plot(data, label='Train Loss')
data = load(f'current_model/alex_sgd_0.01_val_loss.txt')
ax.plot(data, label='Valid Loss')
ax.set_title('Loss over epochs')
ax.set_xlabel('Epoch')
ax.set_ylabel('MSE')
ax.legend()
plt.show()
```
